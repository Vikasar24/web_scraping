```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from html_table_parser.parser import HTMLTableParser
url = "https://kn.wikipedia.org/wiki/%E0%B2%95%E0%B2%A8%E0%B3%8D%E0%B2%A8%E0%B2%A1"
import collections
collections.Callable = collections.abc.Callable
from six.moves import urllib
temp_file, _ = urllib.request.urlretrieve(url)
from urllib.request import urlopen
```

```python
 #Send an HTTP request to the URL
response = requests.get(url)

def url_get_contents(url):


	#making request to the website
	req = urllib.request.Request(url=url)
	f = urllib.request.urlopen(req)

	#reading contents of the website
	return f.read()
```

```python
# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')
```

```python
 # Get the page title
title = soup.title.string
print(f"Title: {title}")
```

```python
 # Get the number of words in the page
words = soup.get_text().split()
num_words = len(words)
print(f"Number of words: {num_words}")
```

```python
 #Get the number of letters in the page (excluding spaces)
num_letters = len(''.join(words))
print(f"Number of letters: {num_letters}")
```

```python
# Get the links in the page
links = [a['href'] for a in soup.find_all('a', href=True)]
print(f"Number of links: {len(links)}")
```

```python
paragraphs = soup.find_all("p")
for paragraph in paragraphs:
    print(paragraph.text)
    
# Extract text from an element with the class "my-class"
elements_with_class = soup.find_all("dl",class_="my-class")
for element in elements_with_class:
    print(element.text)

```

```python
urls = []
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
# Extract images
images = soup.find_all('img')
for img in images:
    print(img['src'])
```

```python
# Find all table elements
tables = soup.find_all('table')

# Get the number of tables
num_tables = len(tables)

# Print the number of tables
print("Number of tables: ", num_tables)
```

```python
# defining the html contents of a URL.
xhtml = url_get_contents("https://kn.wikipedia.org/wiki/%E0%B2%95%E0%B2%A8%E0%B3%8D%E0%B2%A8%E0%B2%A1").decode('utf-8')

# Defining the HTMLTableParser object
p = HTMLTableParser()

# feeding the html contents in the
# HTMLTableParser object
p.feed(xhtml)

# Now finally obtaining the data of
# the table required
print(p.tables[2])

# converting the parsed data to
# dataframe
print("\n\nPANDAS DATAFRAME\n")
df_data=pd.DataFrame(p.tables[2])
print(df_data)

df_data.to_csv('tables output.csv',index=False)
```
